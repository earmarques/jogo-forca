# Jogo da Forca

```python

# ****************************************
#   autor: Éder Marques                  *
#   e-mail: earmarques@gmail.com         *
#   05/22/2020  Gastão Vidigal/SP        *
# ****************************************

# Cabeçário --------------------------------------------------------------------

print("\033[1;30m  Jogo da Forca  ".center(80,"_"))
print()


# Entrada de Dados  ------------------------------------------------------------ 

palavra = input("Entre com a palavra secreta:\n>> ")
palavra = list(palavra.upper())

# Dimensionando lista de chutes e inicializando com espaços
letras_descobertas = []
for c in palavra:
  letras_descobertas.append(' ')


# ATARI Computação Gráfica LTDA-íssima -----------------------------------------

# Paleta de Cores
RESET_COR =        "\033[0;30m"
PRETO =            "\033[1;30m"
VERMELHO =         "\033[1;31m"
VERMELHO_ITALICO = "\033[3;31m" 
AMARELO =          "\033[1;33m"
VERDE =            "\033[1;32m"
AZUL =             "\033[1;34m"
ROXO =             "\033[1;35m"

# Componentes   
trave_superior = " _______"
poste =          "\n|"
corda =          "\t|"
forca =          "       §"
cabeca =         "       @"
cabeca_brava =   "     (`~´)"
cabeca_morta =   "     (x.x)"
corpo =          "      [E]"  # super E'der
braco_direito =        "/"
braco_esquerdo =        "\\"
pernas =         "      / \\"


# Montando a Estrutura ---------------------------------------------------------

linhas = list()
linhas.append('\n\n\n\n\n\n\n\n\n\n\n\n\n\n')   # linha 0:  esconde palavra
linhas.append(list(trave_superior))             # linha 1
linhas.append(list(poste + corda))              # linha 2
linhas.append(list(poste + corda))              # linha 3
linhas.append(list(poste + forca))              # linha 4:  cabecas
linhas.append([])                               # linha 5:  corpo&bracos
linhas.append([])                               # linha 6:  pernas
linhas.append(list(poste))                      # linha 7
linhas.append(list(poste))                      # linha 8
linhas.append([])                               # linha 9:  letras descobertas
linhas.append([])                               # linha 10: base tracejada

# Controles
morreu = False
ganhou = False
erro = 0

while not morreu:

  # linha 4 - cabecas 
  if erro == 6:   # morreu
    linhas[4] = list(poste + cabeca_morta)
  elif erro == 5: # tá quase 
    linhas[4] = list(poste + cabeca_brava)
  elif erro > 0:
    linhas[4] = list(poste + cabeca)
  else:
    linhas[4] = list(poste + forca)

  # linha 5 - corpo&bracos
  if erro > 1:         
    linhas[5] = list(poste + corpo)
    if erro > 2:
      linhas[5][7] = braco_direito 
      linhas[5].append(braco_esquerdo)
      # linha 6 - pernas
      if erro > 3:         
        linhas[6] = list(poste + pernas)          
      else:
        # sem pernas
        linhas[6] = list(poste)
  else:
    # sem corpo&bracos e sem pernas
    linhas[5] = list(poste)

  # Letras Descobertas
  linhas[9] = list(poste)
  for letra in letras_descobertas:
    linhas[9].append("  " + letra.upper() + " ")

  # Base Trasejada
  traco = "___"
  linhas[10] = list(poste)
  for c in palavra:
    linhas[10].extend(list(" " + traco))


  # Colorindo o Desenho --------------------------------------------------------

  linhas[1].insert(0, PRETO)
  linhas[2].insert(0, PRETO)
  linhas[3].insert(0, PRETO)
  
  # Cabeças
  linhas[4].insert(0, PRETO)
  if erro > 0 and erro < 5:
    linhas[4].insert(3, ROXO)
  elif erro == 5:
    linhas[4].insert(3, VERMELHO)
  elif erro == 6:
    linhas[4].insert(3, VERMELHO_ITALICO)
    linhas[4].append(RESET_COR)
  
  # Braços na mesma cor da cabeça  
  linhas[5].insert(0, PRETO)
  if erro > 2 and erro < 5:
    linhas[5].insert(8, ROXO)      # braco
    linhas[5].insert(10, VERDE)    # corpo
    linhas[5].insert(14, ROXO)     # braco
  elif erro >= 5:
    linhas[5].insert(8, VERMELHO)  # braco
    linhas[5].insert(10, VERDE)    # corpo
    linhas[5].insert(14, VERMELHO) # braco
  else:
    # pinta só o corpinho
    linhas[5].insert(3, VERDE)
  
  linhas[6].insert(0, PRETO)  
  linhas[6].insert(3, AZUL)   # pernas
  linhas[7].insert(0, PRETO)  
  linhas[8].insert(0, PRETO)  
  linhas[9].insert(0, PRETO)  
  linhas[9].insert(3, VERDE)  # letras descobertas
  linhas[10].insert(0, PRETO)  


# Renderização  ----------------------------------------------------------------
  
  pincel = []
  for linha in linhas:
    pincel.append("".join(linha))
  
  print("".join(pincel))


# Interação com o Jogador    ---------------------------------------------------

  # Controle
  if ganhou:
    break   # saia do loop
  
  # Vidas do Jogador
  msg = "\n"
  if erro == 6: # game-over
    morreu = True
  elif erro == 5:
    msg += VERMELHO + "  Última chance!!!  ".center(30,"+") + RESET_COR
  else:
    msg += "Você tem {} chances.".format(6 - erro)
  print(msg)

  # Coleta do Chute/Palpite
  if not morreu:
    chute = input("\nArrisque uma letra:\n>> ")
    chute = chute.upper()
    # Validação do chute
    acertou = False
    for indice, letra in enumerate(palavra):    
      if chute == letra:
        acertou = True
        letras_descobertas[indice] = letra
    if not acertou: 
      erro += 1

  # Checando Vencedor
  palavra_secreta = "".join(palavra)
  palavra_descoberta = "".join(letras_descobertas)

  if (palavra_secreta == palavra_descoberta):
    ganhou = True


# Resultado do Jogo ------------------------------------------------------------

if morreu:
  print(VERMELHO + "..... Perdeu Playboy!!!")  
elif ganhou:
  print("\n\n" + VERDE + "  Congratulations!!!  ".center(32, "*"))  

print("\n\n\n\n\n\n")  



```
